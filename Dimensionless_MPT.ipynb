Scripts to generate dimensionless material property targets in the supporting information of  _**Material property targets to enable adsorptive water treatment and resource recovery systems**_

#### Load  libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import multiscale_adsorption_target_models as mat # python model and workflow functions
```

#### Parameters for sensitivity analysis over dimensionless material property target curves

```python
rec_span = np.array([1.5, 5, 10, 50, 100, 1000]) # [-]
eps_span = np.array([0.1, 0.2, 0.3, 0.4]) # [-]
N_BV_span = np.array([50, 500, 1000, 5000, 10000, 50000]) # [-]

K_span = np.logspace(-4,4,500)
```

#### Perform sensitivity analysis over dimensionless MPT curves
This code generates figures S8 - S31 in the supporting information of the manuscript

```python
for i in range(len(rec_span)):
    for j in range(len(eps_span)):
        print('Calculating dimensionless targets for r={0} eps={1}'.format(rec_span[i],eps_span[j]))
        
        figname = 'nd_mpt_r{0}_eps{1}.png'.format(rec_span[i], eps_span[j])
        
        QC, QB, xKQ = mat.nondim_sens(rec_span[i],eps_span[j],N_BV_span,K_span)
        
        fig, ax = mat.plot_dimensionless_targets(K_span, QC, QB, xKQ, N_BV_span, rec_span[i], eps_span[j], save=figname)
    # End
# End        
```
