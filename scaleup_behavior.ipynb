Scripts to generate plots for the scale up behavior of batch and semicontinuous processes in  _**Material property targets to enable adsorptive water treatment and resource recovery systems**_

#### Load  libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import multiscale_adsorption_target_models as mat # python model and workflow functions
```

#### Parameters for Type 1 and Type 2 behavior

```python
rec_span = np.array([1.1, 10]) # [-]
eps_span = np.array([0.3]) # [-]
N_BV_span = np.array([10, 500, 50000]) # [-]

K_span = np.logspace(-4,4,500)
```

#### Plot Type 1 and Type 2 behavior
This code generates Fig 5A, 5B of the manuscript

```python
for i in range(len(rec_span)):
    for j in range(len(eps_span)):
        print('Calculating dimensionless targets for r={0} eps={1}'.format(rec_span[i],eps_span[j]))
        
        figname = 'scale_up_type{0}.png'.format(i+1)
        
        QC, QB, xKQ = mat.nondim_sens(rec_span[i],eps_span[j],N_BV_span,K_span)
        
        fig, ax = mat.plot_dimensionless_targets(K_span, QC, QB, xKQ, N_BV_span, rec_span[i], eps_span[j], save=figname)
    # End
# End        
```

### Parameters for Type 3 behavior
